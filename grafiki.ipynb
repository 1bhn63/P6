Шаг 1. Импортируем pandas для работы с датасетом и matplotlib для работы с графиками

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
titanic = pd.read_csv('titanic.csv')
titanic
```

Шаг 2. Сделаем 3 штуки линейного графика

```python
plt.figure(figsize=(15, 5))
plt. plot(titanic['Name'].head(10), titanic['Sex'].head(10), label='пол')
plt.title('Пол персонажей')
plt.xlabel('Имя персонажа')
plt.ylabel('Пол')
plt.legend()
plt.show()
```

Был сделан линейный график

```python
plt.figure(figsize=(15, 5))
plt. plot(titanic['Name'].head(10), titanic['Age'].head(10), label='возраст', marker = 'o', markersize = 7)
plt.title('Возраст персонажей')
plt.xlabel('Имя персонажа')
plt.ylabel('Возраст')
plt.legend()
plt.show()
```

Был создан линейный график

```python
plt.figure(figsize=(15, 5))
plt. plot(titanic['Name'].head(10), titanic['Fare'].head(10), label='оплата', marker = 'o', markersize = 7, color = 'pink')
plt.title('оплата персонажей')
plt.xlabel('Имя персонажа')
plt.ylabel('оплата')
plt.legend()
plt.grid()
plt.show()
```

Шаг 3. Сделаем 3 штуки стобчатой диаграммы 


```python
plt.figure(figsize=(25, 10))
plt.bar(titanic['Name'].head(11), titanic['Fare'].head(11), label='Оплата', color = "orange")
plt.xlabel('Имя персонажа')
plt.ylabel('Оплата')
plt.title('Оплата персонажей')
plt. legend( )
plt.show()
```

Siblings/Spouses Aboard	Parents/Children Aboard

```python
plt.figure(figsize=(25, 10))
plt.bar(titanic['Name'].head(11), titanic['Siblings/Spouses Aboard'].head(11), label='Супруги/дети', color = "green")
plt.bar(titanic['Name'].head(11), titanic['Parents/Children Aboard'].head(11), label='Семья', color = "red")
plt.xlabel('Имя персонажа')
plt.title('Семья')
plt. legend( )
plt.show()
```

```python
plt.figure(figsize=(25, 10))
plt.bar(titanic['Name'].tail(11), titanic['Age'].tail(11), label='ВОзраст', color = "yellow")
plt.xlabel('Имя персонажа')
plt.ylabel('Возраст')
plt.title('Возраст персонажей')
plt. legend( )
plt.show()
```

Шаг 4. Сделаем 2 штуки круговых диаграмм 


```python
plt.figure(figsize=(25, 10))
vozrast = titanic['Age'].value_counts()
plt.pie(vozrast, labels= vozrast.index, autopct='%1.lf%%',
colors=['purple', 'blue', 'orange', 'pink', 'red', 'green', 'grey', 'black', 'White', 'gold', 'brown', 'cyan'])
plt.title('Соотношение возраста')
plt.show()
```

```python
plt.figure(figsize=(25, 10))
oplata = titanic['Fare'].value_counts()
plt.pie(oplata, labels= oplata.index, autopct='%1.lf%%',
colors=['purple', 'blue', 'orange', 'pink', 'red', 'green', 'grey', 'black', 'White', 'gold', 'brown', 'cyan'])
plt.title('Соотношение оплаты')
plt.show()
```

Шаг 5. Сделаем  2 диаграммы рассеяния


```python
plt. figure(figsize=(5, 6))
plt.scatter(titanic['Sex'].head(50), titanic['Age'].head(50), color='purple')
plt.title('Связь между возрастом и полом у первых 50 персонажей')
plt.xlabel('пол')
plt.ylabel('возраст')
plt.show()
```

```python
plt. figure(figsize=(5, 6))
plt.scatter(titanic['Siblings/Spouses Aboard'].head(50), titanic['Parents/Children Aboard'].head(50), color='cyan')
plt.title('Связь ')
plt.xlabel('Братья сёстры')
plt.ylabel('Супруги дети')
plt.show()
```

Шаг 6. Сделаем 3 гистограммы

```python
plt.figure(figsize=(10, 6))
plt.hist(titanic['Age'], bins=30, color=["cyan"], edgecolor='lime', alpha=0.7, label = 'Распределение Возраста персонажей')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(titanic['Sex'], bins=30, color=["cyan"], edgecolor='lime', alpha=0.7, label = 'Распределение пола персонажей')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(titanic['Fare'], bins=30, color=["cyan"], edgecolor='lime', alpha=0.7, label = 'Распределение оплаты персонажей')
plt.legend()
plt.show()
```
